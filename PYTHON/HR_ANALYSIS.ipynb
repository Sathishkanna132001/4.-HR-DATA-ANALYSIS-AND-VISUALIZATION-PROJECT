# QUESTION : 1 Count number of attrition "Yes" and "No" in each department?

```python
Attrition_count = hr.groupby(["Department","Attrition"]).size().unstack(fill_value=0)
print(Attrition_count)
```

# DISPLAY STACKED BAR CHART:

```python
import matplotlib.pyplot as plt

Attrition=Attrition_count.plot(kind="bar",stacked=True,color=["#FF0000","#00B050"])

plt.xlabel('Department')
plt.ylabel('Count')
plt.title('Count of Attrition ("Yes" and "No") in Each Department')
plt.legend(title='Attrition', labels=['No','Yes'])
plt.xticks(rotation=45) 
plt.tight_layout()
plt.show()
```

# QUESTION : 2 Monthly Average Income and Hike % By Department?	

```python
monthly_avg_income = hr.groupby('Department')['Monthly_Income'].mean()

avg_hike_percentage = hr.groupby('Department')['Percent_Salary_Hike'].mean()

combined_data = pd.DataFrame({'Monthly_Avg_Income': monthly_avg_income,'Avg_Hike_Percentage': avg_hike_percentage})

print(combined_data)
```

# DISPLAY BY LINE AND CLUSTERED CHART:

```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots(figsize=(12, 8))

bar_width = 0.4
bar_positions = combined_data.index

ax1.bar(bar_positions, combined_data['Monthly_Avg_Income'], width=bar_width, color='b', label='Monthly Avg Income')

ax1.set_xlabel('Department')
ax1.set_ylabel('Monthly Avg Income', color='b')
ax1.tick_params(axis='y', labelcolor='b')

ax2 = ax1.twinx()
ax2.plot(bar_positions, combined_data['Avg_Hike_Percentage'], color='r', marker='o', linestyle='-', linewidth=2, label='Avg Hike Percentage')
ax2.set_ylabel('Avg Hike Percentage (%)', color='r')
ax2.tick_params(axis='y', labelcolor='r')

plt.title('Monthly Average Income and Hike Percentage by Department')

ax1.set_xticks(bar_positions)
ax1.set_xticklabels(combined_data.index, rotation=45, ha='right')

ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()
```

# Question : 3 What is the monthly average income of Educational Field ?						

```python
education_field = hr.groupby('Education_Field')['Monthly_Income'].mean().reset_index()

education_field.columns = ['Education_Field', 'Average_Monthly_Income']

print(education_field)
```

# DISPLAY BY LINE CHART:

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.lineplot(data=education_field, x='Education_Field', y='Average_Monthly_Income', marker='o', color='b')

plt.xlabel('Education Field')
plt.ylabel('Average Monthly Income')
plt.title('Monthly Average Income by Education Field')

plt.xticks(rotation=45)

for i, value in enumerate(education_field['Average_Monthly_Income']):
    plt.text(i, value, f'{value:.2f}', ha='center', va='bottom', color='black', fontweight='bold')

plt.tight_layout()
plt.show()
```
